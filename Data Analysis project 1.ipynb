```python
# Imporing packages for data preparation
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
df = pd.read_csv(r'C:\Users\ALEX\Desktop\CSV SQL\data.csv')
```

```python
# Read all data
pd.options.display.max_rows = 9999
print(df)
```

```python
# Using mean value imputation method to replace all null values by mean on the Calories column 
d = df["Calories"].mean()
df["Calories"].fillna(d, inplace = True)
print(df)
```

```python
# Creating a graph by comparing Duration and Calories
x = df["Duration"].values.reshape(-1, 1)
y = df["Calories"].values.reshape(-1, 1)

linear_regressor = LinearRegression()
linear_regressor.fit(x,y)
y_pred = linear_regressor.predict(x)

plt.scatter(x,y)
plt.plot(x, y_pred, color = "red")
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title("Calories burnt by time exercising")
plt.show()
```

```python
# Using R Squared to see correlation between the variables
R_square = r2_score(y,y_pred)
print(R_square)
```
